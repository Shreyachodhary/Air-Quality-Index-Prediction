**Handling the missing values**

```python
air_quality_data=air_quality_data.replace(to_replace=-200,value=np.nan)
```

```python
air_quality_data.mean(numeric_only=True)
```

```python
#replacing the missing values with mean value
air_quality_data=air_quality_data.fillna(air_quality_data.mean(numeric_only=True))
```

**Forecasting with FB Prophet Algorithm**

FB Prophet documentation:https://facebook.github.io/prophet/docs/quick_start.html

```python
# converting the data from DD/MM/YYYY to YYYY-MM-DD
date_info=pd.to_datetime(air_quality_data['Date'],dayfirst=True)
print(date_info)
```

```python
time_info=air_quality_data['Time']
```

```python
time_info=time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining 2 series to a pandas dataframe
date_time=pd.concat([date_info,time_info],axis=1)
```

```python
date_time.head()
```

```python
date_time.shape
```

```python
# combining date and time
date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

We have to convert ds from object datatype to date time format.

```python
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['NOx(GT)']
```

```python
# Clean negative values (air pollutants cannot be negative)
data['y'] = data['y'].clip(lower=0)

# Handle missing values
df = data.fillna(method='ffill')
```

```python
data.head()
```

```python
df['y'] = np.log1p(df['y'])   # log-transform to keep values positive
```

```python
! pip install prophet
```

```python
from prophet import Prophet
```

```python
#training the Prophet model

model1=Prophet()
model1.fit(data)
```

```python
future=model1.make_future_dataframe(periods=365,freq='H')
future.tail()
```

periods = 365*24 *20
→ That’s 365 days × 24 hours × 20 years = 175,200 periods.
So Prophet will generate 175,200 future hourly timestamps after the end of your training data.

freq = 'H'
→ Each step is 1 hour.

```python
forecast=model1.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1=model1.plot(forecast)
```

```python
fig2=model1.plot_components(forecast)
```
